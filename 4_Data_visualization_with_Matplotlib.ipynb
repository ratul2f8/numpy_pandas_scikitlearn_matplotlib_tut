## Matplotlib example workflow

```python
# 0: import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

#1: Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

#2: Set up plot
fig, ax = plt.subplots(figsize = (10, 10)) # figsize(width, height)

#3: plot the data
ax.plot(x, y)

#4: Customize/Label the plot
ax.set(title = "Simple Plot", xlabel = "x-axis", ylabel = "y-axis")

#5: Save and show the figure
fig.savefig("images/sample-plot.png")
```

## Craete data with numpy array
We want
* Line plot
* Scatter plot
* Bar plot
* Histogram
* Subplots

```python
import numpy as np
# create some data
x = np.linspace(0, 10, 100)
x[:5]
```

```python
# plot the data
fig, ax = plt.subplots()
ax.plot(x, x**2)
```

```python
# scatter
fig, ax = plt.subplots(figsize = (10, 10))
ax.scatter(x, np.exp(x))
```

```python
# scatter
fig, ax = plt.subplots(figsize = (10, 10))
ax.scatter(x, np.sin(x))
```

```python
# make a bar plot from dictionary
nut_butter_prices = {"Almond Butter": 10, "Peanut Butter": 8, "Cashew Butter" : 12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
```

```python
# plot horizontally
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()))
```

```python
# drawing histogram
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x)
```

```python
# subplots option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows = 2,
                                            ncols = 2,
                                            figsize = (10, 5))
ax1.plot(x, x/2)
ax2.scatter(x, np.exp(x))
ax3.scatter(x, np.sin(x))
ax4.hist(x)
```

```python
# subplots - option 2
fig, ax= plt.subplots(nrows = 2,
                                            ncols = 2,
                                            figsize = (10, 5))
ax[0,0].plot(x, x/2)
ax[0,1].scatter(x, np.sin(x))
ax[1,0].hist(x)
```

```python
ts = pd.Series(np.random.randn(1000), index = pd.date_range("1/1/2020", periods = 1000))
ts = ts.cumsum()
ts.plot()
```

```python
car_sales = pd.read_csv("car-sales.csv")
car_sales["Price"] = car_sales["Price"].str.replace('[\$\,\.]', '')
car_sales["Price"] = car_sales["Price"].str[:-2]
number_of_rows = car_sales.shape[0] # or number_of_rows = len(car_sales)
car_sales["Sale Date"] = pd.date_range("1/1/2020", periods= number_of_rows)
car_sales
```

```python
# convert string to integer and make a cumsum
car_sales["Total Sales"] = car_sales["Price"].astype(int).cumsum()
car_sales
```

```python
# show the data
car_sales.plot(x = "Sale Date", y = "Total Sales");
```

```python
car_sales["Odometer (KM)"] = car_sales["Odometer (KM)"].astype(int)
car_sales.plot(x = "Odometer (KM)", y = "Price", kind ="scatter")
```

```python
# another dataset
heart_disease = pd.read_csv("heart-disease.csv")
heart_disease.head()
```

```python
heart_disease.plot()
```

```python
heart_disease.plot.hist(figsize = (10, 30),subplots = True)
```

```python
over_50 = heart_disease[heart_disease["age"] > 50 ]
over_50
```

```python
over_50.plot(x = "age", y = "chol",kind = "scatter", c = "target")
```

```python
## the OO method
fig, ax = plt.subplots(figsize = (10, 6))
over_50.plot(kind = 'scatter', x = "age", y = "chol", ax = ax);
```

```python
fig, ax = plt.subplots(figsize = (10, 6))
over_50.plot(kind = 'scatter', x = "age", y = "chol", ax = ax)
ax.set_xlim([45, 100])
```

```python
# another way of writing the OO method
fig, ax = plt.subplots(figsize = (10, 6))
scatter = ax.scatter(x = over_50["age"], y = over_50["chol"], c = over_50["target"])
ax.set( title = "Heart disease and cholesterol level", xlabel = "Age", ylabel = "Cholesterol")
# add a legend
ax.legend(*scatter.legend_elements(), title = "Target");

# add a horizontal line
ax.axhline(over_50["chol"].mean(), ls = "--") # "ls" can be written as linestyle
```

```python
fig, (ax0, ax1) = plt.subplots(figsize = (10, 6), nrows = 2, ncols = 1)
scatter = ax0.scatter(x = over_50["age"], y = over_50["chol"], c = over_50["target"])
ax0.set( title = "Heart disease and cholesterol level", xlabel = "Age", ylabel = "Cholesterol")
# add a legend
ax0.legend(*scatter.legend_elements(), title = "Target");

# add a horizontal line
ax0.axhline(over_50["chol"].mean(), ls = "--") # "ls" can be written as linestyle
scatter = ax1.scatter(x = over_50["age"], y = over_50["thalach"], c = over_50["target"])
ax1.set( title = "Heart disease and Max Heart rate ", xlabel = "Age", ylabel = "Max Heart Rate")
# add a legend
ax1.legend(*scatter.legend_elements(), title = "Target");

# add a horizontal line
ax1.axhline(over_50["thalach"].mean(), ls = "--") # "ls" can be written as linestyle
```

```python
#sharex
fig, (ax0, ax1) = plt.subplots(figsize = (10, 6), nrows = 2, ncols = 1)
scatter = ax0.scatter(x = over_50["age"], y = over_50["chol"], c = over_50["target"])
ax0.set( title = "Heart disease and cholesterol level", ylabel = "Cholesterol")
# add a legend
ax0.legend(*scatter.legend_elements(), title = "Target");

# add a horizontal line
ax0.axhline(over_50["chol"].mean(), ls = "--") # "ls" can be written as linestyle
scatter = ax1.scatter(x = over_50["age"], y = over_50["thalach"], c = over_50["target"])
ax1.set( title = "Heart disease and Max Heart rate ", xlabel = "Age", ylabel = "Max Heart Rate")
# add a legend
ax1.legend(*scatter.legend_elements(), title = "Target");

# add a horizontal line
ax1.axhline(over_50["thalach"].mean(), ls = "--") # "ls" can be written as linestyle

#add super title
fig.suptitle("Heart Disease Analysis", fontsize = 16, fontweight = "bold")
```

## Customizing as well as styling the plots of matplotlib

```python
# see the different available styles
plt.style.available
```

```python
car_sales = pd.read_csv("car-sales.csv")
car_sales["Price"] =  car_sales["Price"].str.replace('[\$\,\.]', '')
car_sales["Price"] =  car_sales["Price"].str[:-2]
car_sales["Price"] =  car_sales["Price"].astype(int)
#apply the style
plt.style.use('seaborn-whitegrid')
car_sales["Price"].plot()
```

```python
df = pd.DataFrame(np.random.randn(10,4), columns=["a", "b", "c", "d"])
ax = df.plot(kind = "bar")
ax.set( title = "Random Number Bar Graph From a Daata Frame")
ax.legend().set_visible(True);
```

```python
#set style
plt.style.use('seaborn-whitegrid')
# another way of writing the OO method
fig, ax = plt.subplots(figsize = (10, 6))
scatter = ax.scatter(x = over_50["age"], 
                     y = over_50["chol"], 
                     c = over_50["target"],
                     cmap = "winter") # here the cmap changes the color scheme
ax.set( title = "Heart disease and cholesterol level",
       xlabel = "Age",
       ylabel = "Cholesterol")
# add a legend
ax.legend(*scatter.legend_elements(), title = "Target");

# add a horizontal line
ax.axhline(over_50["chol"].mean(), ls = "--") # "ls" can be written as linestyle
```

```python
# customizing the y and x axis limitation
plt.style.use('seaborn-whitegrid')
#sharex
fig, (ax0, ax1) = plt.subplots(figsize = (10, 10), nrows = 2, ncols = 1)
scatter = ax0.scatter(x = over_50["age"], y = over_50["chol"], c = over_50["target"], cmap = 'winter')
ax0.set( title = "Heart disease and cholesterol level", ylabel = "Cholesterol")
# add a legend
ax0.legend(*scatter.legend_elements(), title = "Target");

#change the x axis limits
ax0.set_xlim([50, 80])
#change the x axis limits
ax0.set_ylim([100, 600])
# add a horizontal line
ax0.axhline(over_50["chol"].mean(), ls = "--") # "ls" can be written as linestyle
scatter = ax1.scatter(x = over_50["age"], y = over_50["thalach"], c = over_50["target"], cmap = 'plasma')
ax1.set( title = "Heart disease and Max Heart rate ", xlabel = "Age", ylabel = "Max Heart Rate")
# add a legend
ax1.legend(*scatter.legend_elements(), title = "Target");
#change the x axis limits
ax1.set_xlim([50, 80])
#change the y axis limits
ax1.set_ylim([60, 200])
# add a horizontal line
ax1.axhline(over_50["thalach"].mean(), ls = "--") # "ls" can be written as linestyle

#add super title
fig.suptitle("Heart Disease Analysis", fontsize = 16, fontweight = "bold")
```
